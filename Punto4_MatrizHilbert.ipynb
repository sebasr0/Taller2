```python
#
import numpy as np
import matplotlib.pyplot as plt

# 1. Matriz de Hilbert
def hilbert(n):
    H = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            H[i,j] = 1 / (i + j + 1)
    return H

# 2. Simular datos normales    
def simulate(n):
    H = hilbert(n)
    return np.random. multivariate_normal(np.zeros(n), H, size=1000)

# 3. Estimar matriz de covarianza
def estimate_cov(X):
    return np.cov(X.T)

# 4. Graficar número de condición
ns = range(5, 100, 5)
conds = []

for n in ns:
    X = simulate(n)
    C = estimate_cov(X)
    conds.append(np.linalg.cond(C))
    
plt.plot(ns, conds)
plt.xlabel('n')
plt.ylabel('Número de condición')
```
