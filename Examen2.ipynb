## Graficando con respecto a n

```python
def matriz_potencias(n):
    """
    Crea una matriz de potencias de tamaño nxn.
    """
    A = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            A[i, j] = (i+1)**(j+1)
    return A

# Lista para almacenar números de condición
condiciones = []

# Rango de tamaños de matriz para evaluar
n_values = list(range(1, 80))

for n in n_values:
    A = matriz_potencias(n)
    # Calcular el número de condición usando numpy
    condicion = np.linalg.cond(A)
    condiciones.append(condicion)

# Graficar
plt.figure(figsize=(10,6))
plt.plot(n_values, condiciones, '-o')
plt.yscale('log')
plt.title('Número de condición de la matriz de potencias')
plt.xlabel('Tamaño de la matriz (n)')
plt.ylabel('Número de condición (escala logarítmica)')
plt.grid(True, which="both", ls="--", c='0.65')
plt.show()
```
